#### 6 Sentence Statistics

```python
import jieba
def tokenize_chinese(text):
    l = list(jieba.cut(text))
    return [ i for i in l if i != " " and not isinstance(i, int)]

def count_token(arr):
    return len(arr)


tokenize_chineseDF = udf(lambda x : tokenize_chinese(x), ArrayType(StringType()))
```

```python
from pyspark.sql.functions import size

bi = bi.withColumn("token", tokenize_chineseDF(bi.text))
bc = bc.withColumn("token", tokenize_chineseDF(bc.text))
bm = bm.withColumn("token", tokenize_chineseDF(bm.text))

bi = bi.repartition(60)
bc = bc.repartition(60)
bm = bm.repartition(60)

import shutil
shutil.rmtree("output/6_BC_STATS.avro")


write_avro(bc.withColumn("length", size(bc.token)), "output/6_BC_STATS.avro")
write_avro(bm.withColumn("length", size(bm.token)), "output/6_BM_STATS.avro")
write_avro(bi.withColumn("length", size(bi.token)), "output/6_BI_STATS.avro")
```

### Filter outlier and get small sample

```python
from utilities import find_outliers
```

#### FILTER ENG OUTLIER

```python
filBI = read_avro("output/6_BI_STATS.avro")
biC = filBI.select(filBI.length)
biC.toPandas().plot.box()  
obi = find_outliers(filBI)
nobi = obi.filter((obi.total_outliers != 1)  & (obi.length > 5))
biC = nobi.select(nobi.length)
biC.toPandas().plot.box()
nobi.count()
```

```python
nobi.drop("token").drop("total_outliers").drop("lang").drop("length").sample(0.48, 1).write.mode("overwrite").csv("output/7_biReadyAnalysis_B.csv")
```
