```python
from connectPyspark import *
from pyspark.sql.window import Window
import pyspark.sql.functions as F
from pyspark.sql.functions import row_number
from pyspark.sql.functions import desc
from pyspark.sql.functions import array_contains
import utilities as u
```

```python
bc = spark.read.format("avro").load("output/bcwc.avro")
bm = spark.read.format("avro").load("output/bmwc.avro")
bi = spark.read.format("avro").load("output/biwc.avro")
```

```python
bc.select("word","word_count", F.row_number().over(Window.partitionBy() \
  .orderBy(desc(bc['word_count']))).alias("id")).show(1000)
```
